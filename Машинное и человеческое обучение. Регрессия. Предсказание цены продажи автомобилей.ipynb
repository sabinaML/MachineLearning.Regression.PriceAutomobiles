```python
def print_error(col):
    b=testm.groupby(col).agg({'error_abs':['count', 'mean'], 'error':['count', 'mean']})
    b.columns=['_'.join(col).strip() for col in b.columns.values]
    b=b.drop('error_count', axis=1)
    b['mean_error_diff']=b['error_abs_mean']-testm['error_abs'].mean()
    b['mean_error'] = testm['error_abs'].mean()
    display(b)
```
